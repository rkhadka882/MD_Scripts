```python
import numpy as np
import os
import sys
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams.update({'font.serif':'Arial', 
                     'font.size': 14.0,
                     'font.stretch':'normal',
                     'xtick.labelsize': 12, 
                     'ytick.labelsize': 12,
                     'legend.fontsize':12,
                     'legend.shadow':'False',
                     'figure.dpi': 80
                    })
```

```python
# muCo: -0.6, muSi = -1.5
file = './Log_files/chemical_potential.dat'
fig = plt.figure(figsize = (4,3))
mu_data = pd.read_csv(file, header=0, delimiter=' ')
print(mu_data.columns, '\n')

# Pressures
plt.plot(mu_data['Time'], mu_data['pressure'], '-k')
plt.xlabel('Time [ps]')
plt.ylabel('Pressure [bar]')
# plt.title('Pressure vs Time')
plt.axhline(0, linestyle='dashed', color='k')
plt.savefig('./Log_files/Pressure.png', dpi=300, bbox_inches = 'tight', facecolor="w")
plt.show()

fig = plt.figure(figsize = (4,3))
# Density
plt.plot(mu_data['Time'], mu_data['density'], '-b')
plt.xlabel('Time [ps]')
plt.ylabel('Density [$gm/cm^3$]')
# plt.title('Density vs Time')
plt.axhline(6.8, linestyle='dashed', color='k')
plt.savefig('./Log_files/Density.png', dpi=300, bbox_inches = 'tight', facecolor="w")
plt.show()

# Stoichiometry and number of particles:
file = './Log_files/GCMC_CoSi_proba_stat.dat'
GCMC_data = pd.read_csv(file, header=0, delimiter=' ')
print(GCMC_data.columns, '\n')

fig = plt.figure(figsize = (4,3))
# Stoichiometry of atoms
plt.plot(0.001*GCMC_data['TimeStep'], GCMC_data['v_ratio'], '-r')
plt.xlabel('Time [ps]')
plt.ylabel('Stoichiometry')
# plt.title('Stoichiometry vs Time')
plt.ylim(0.9, 1.3)
# plt.axhline(6.8, linestyle='dashed', color='k')
plt.savefig('./Log_files/Stoichiometry.png', dpi=300, bbox_inches = 'tight', facecolor="w")
plt.show()

fig = plt.figure(figsize = (4,3))
plt.plot(0.001*GCMC_data['TimeStep'], GCMC_data['v_nCo'], label='$n_{Co}$')
plt.plot(0.001*GCMC_data['TimeStep'], GCMC_data['v_nSi'], label='$n_{Si}$')
plt.xlabel('Time [ps]')
plt.ylabel('Number of atoms')
# plt.title('Atom count vs Time')
# plt.ylim(0.9, 1.3)
plt.legend()
# plt.axhline(6.8, linestyle='dashed', color='k')
plt.savefig('./Log_files/Atom_count.png', dpi=300, bbox_inches = 'tight', facecolor="w")
plt.show()

# Probability
fig = plt.figure(figsize = (4,3))
plt.plot(0.001*GCMC_data['TimeStep'], GCMC_data['v_Co_iprob'], label='$P(Co)$')
plt.plot(0.001*GCMC_data['TimeStep'], GCMC_data['v_Si_iprob'], label='$P(Si)$')
plt.xlabel('Time [ps]')
plt.ylabel('Atom insertion probability')
# plt.title('Atom count vs Time')
# plt.ylim(0.9, 1.3)
plt.legend()
# plt.axhline(6.8, linestyle='dashed', color='k')
plt.savefig('./Log_files/Atom_insertion_prob.png', dpi=300, bbox_inches = 'tight', facecolor="w")
plt.show()

```
