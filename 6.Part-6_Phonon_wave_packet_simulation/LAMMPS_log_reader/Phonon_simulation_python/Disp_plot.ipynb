```python
import numpy as np
import sys
import os
import seaborn as sns
import matplotlib.pyplot as plt
import cmath
```

```python
# Value of variables:
A = 0.15                         # range from 0.015 - 0.15 A
eps_k = 1                        # Polarization vector
a = 4.0782                       # a = lattice parameter, in A

k0 = 0.1*np.pi/a                 # wave vector
w = 3.6                          # frequency corresponding to k0, in THz
x0 = 1300                        # middle of the Au layer, for 2600 A
xn = np.linspace(0, 2600, 962)  # Total distance in A

zeta = 50*a                      # spatial extent
y = 0 + k0*(xn - x0)*1j

disp = A*eps_k*np.exp(-(xn - x0)**2/zeta**2)*np.exp(y)

# Plot the position
U = np.real(disp)
plt.plot(xn, U)
plt.xlabel("Z-axis coordinate [A]")
plt.ylabel("Displacement [A]")
# plt.xlim(800, 1800)
# plt.ylim(-0.3, 0.3)
plt.show()


# Plot the velocity
vz = (w*1e12*np.imag(disp))/1e15
plt.plot(xn, vz)
plt.xlabel("Z-axis coordinate [A]")
plt.ylabel("velocity, vz [A/fs]")
# plt.xlim(800, 1800)
# plt.ylim(-0.3, 0.3)
plt.show()
```
