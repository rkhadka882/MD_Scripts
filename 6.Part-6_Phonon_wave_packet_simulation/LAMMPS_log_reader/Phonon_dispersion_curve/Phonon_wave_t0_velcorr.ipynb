```python
import numpy as np
import sys
import os
import seaborn as sns
import matplotlib.pyplot as plt
import cmath
```

```python
# os.system('ls')
```

```python
# Reading the phonon dispersion data:
filename = 'Au_morse_GX.disp'
with open(filename, 'r') as file:
    lines = file.readlines()

data = [line.split() for line in lines if not line.startswith('#')]

data = np.array(data)
data = data.astype(np.float64)
cm_2_THz = 29979245800/1e12
TAx = data[0::3][:, 1]*cm_2_THz  # in (cm-1)
TAy = data[1::3][:, 1]*cm_2_THz  # in (cm-1)
LA = data[2::3][:, 1]*cm_2_THz   # in (cm-1)
k_val = (data[0::3][:, 0] - 1)/100

plt.plot(k_val, TAx, 'k', label='TAx mode')
plt.plot(k_val, TAy, 'k', label='TAy mode')
plt.plot(k_val, LA, 'k', label='LA mode')
plt.xlim(0, 2)
plt.ylim(0, 7)
# plt.xlabel('k values [$\pi$/a]')
# plt.ylabel('Frequency [THz]')
# plt.title('Phonon dispersion relation of Au, (000) to (111) ')
# plt.legend()
plt.show()
GX_data = np.hstack((k_val[:, None], LA[:, None], TAx[:, None], TAy[:, None])); print(GX_data.shape)
np.savetxt("0GX_data.dat", GX_data)
```
